```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import re
import requests
```

```python
product_name=[]
distances=[]
emis=[]
#original_prices=[]
final_prices=[]
locations=[]
```

```python
req_header={
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/149.0.0.0 Safari/537.36",
    "Accept-Language": "en-GB,en-US;q=0.9,en;q=0.8",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Referer": "https://www.google.com/"
    }
for i in range(1,1001):
    url=f"https://www.cars24.com/buy-used-car/?sort=bestmatch&serveWarrantyCount=true&listingSource=FilterTags&storeCityId={i}"
    #print(url)
    response=requests.get(url,headers=req_header)
    #print("Page :",i,response.status_code)
    html_data=response.text
    soup=BeautifulSoup(html_data)
    # Extract product name
    product=soup.find_all(class_='sc-ksBlki dxpZAa')
    for i in product:
      product_name.append(i.text)
    #Extract travalled distance
    distance=soup.find_all(class_='sc-ksBlki knJTXq')
    for i in distance:
      distances.append(i.text)
    #Extract emi
    emi=soup.find_all(class_='styles_flexItem__2z__J')
    for i in emi:
      emis.append(i.text)
    #Extract Original Price
    #original_price=soup.find_all(class_='sc-ksBlki fAywVv')
    #for i in original_price:
      #original_prices.append(i.text)
    #Extract Final Price
    final_price=soup.find_all(class_='sc-ksBlki dMXwkk')
    for i in final_price:
      final_prices.append(i.text)
    #Extract location
    location=soup.find_all(class_='styles_ellipsis__uatjG')
    for i in location:
      locations.append(i.text)
```

```python
km = [int(re.sub(r'[^0-9]', '', x)) for x in distances if re.search(r'km$', x)]
fuel = [x for x in distances if re.fullmatch(r'Petrol|Diesel|CNG|Electric|Hybrid', x)]
transmission = [x for x in distances if re.fullmatch(r'Manual|Automatic', x)]
registration = [x for x in distances if re.fullmatch(r'[A-Z]{2}-\d{1,2}', x)]
```

```python
print("Car Name:",product_name)
print("-"*150)
print("Kilometers:",km)
print("-"*150)
print("Fuel:",fuel)
print("-"*150)
print("Transmission:",transmission)
print("-"*150)
print("Registration:",registration)
print("-"*150)
print("EMI:",emis)
print("-"*150)
#print("Original Price:",original_prices)
#print("-"*150)
print("Final Price:",final_prices)
print("-"*150)
print("Location:",locations)
```

```python
print("Car Name:",len(product_name))
print("-"*150)
print("Kilometers:",len(km))
print("-"*150)
print("Fuel:",len(fuel))
print("-"*150)
print("Transmission:",len(transmission))
print("-"*150)
print("Registration:",len(registration))
print("-"*150)
print("EMI:",len(emis))
print("-"*150)
#print("Original Price:",len(original_prices))
#print("-"*150)
print("Final Price:",len(final_prices))
print("-"*150)
print("Location:",len(locations))
```

```python
min_len_list=[
    len(product_name),
    len(emis),
    len(km),
    len(fuel),
    len(transmission),
    len(registration),
    #len(original_prices),
    len(final_prices),
    len(locations) ]
min_len = min(min_len_list)
print(min_len)
df=pd.DataFrame({
    "Car Name":product_name[:min_len],
    "Kilometers":km[:min_len],
    "Fuel":fuel[:min_len],
    "Transmission":transmission[:min_len],
    "Registration":registration[:min_len],
    "EMI":emis[:min_len],
    #"Original Price":original_prices[:min_len],
    "Final Price":final_prices[:min_len],
    "Location":locations[:min_len]
})
df.shape
```
